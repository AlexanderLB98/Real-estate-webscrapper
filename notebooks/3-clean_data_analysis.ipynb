```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.float_format', '{:.2f}'.format)
```

# Processed data Analysis


## Load Data

```python
df = pd.read_csv("../data/processed/processed_data-valencia_barcelona.csv")
df
```

```python
print("\nRango de precios:")
print(f"Mínimo: {df['price'].min()}, Máximo: {df['price'].max()}, Media: {df['price'].mean():.2f}")

# Distribución de precios
plt.figure(figsize=(10, 6))
sns.histplot(df['price'], kde=True, bins=30, color='blue')
plt.title("Distribución de precios")
plt.xlabel("Precio")
plt.ylabel("Frecuencia")
plt.show()
```

```python
# Análisis de metros cuadrados
plt.figure(figsize=(10, 6))
sns.histplot(df['meters'].dropna(), kde=True, bins=30, color='green')
plt.title("Distribución de metros cuadrados")
plt.xlabel("Metros cuadrados")
plt.ylabel("Frecuencia")
plt.show()
```

```python
# Análisis de habitaciones
plt.figure(figsize=(10, 6))
sns.countplot(x='rooms', data=df)
plt.title("Distribución de número de habitaciones")
plt.xlabel("Número de habitaciones")
plt.ylabel("Frecuencia")
plt.show()
```
